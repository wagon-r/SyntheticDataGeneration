```python
import pandas as pd
import numpy as np
```

Reading original data

```python
original_data = pd.read_parquet("../../data/test_107_reduced.parquet")
original_data
```

Defining relevant column types

```python
failure_columns = [
	'cd_counter_rollers_installation',
	'ld_counter_rollers_installation',
	'closing_device_failure',
	'misalignment_cd_vs_ld',
	'misalignment_cd_vs_ld_sill_gap',
	'belt_tension',
	'pulley_is_touching_belt',
	'zero_position',
]

environment_columns = [
	'cmcouplerfriction',
	'cmdoorfriction',
	'cmelectronicage',
	'cmvibration',
]

static_state_columns = [
	'doorcyclecounter',
]

time_state_columns = [
	'doorforce',
	'doorspeed',
	'doorposition',
]

mean_columns = [
	'mean_doorforce',
	'mean_doorspeed',
	'mean_doorposition',
]

median_columns = [
	'median_doorforce',
	'median_doorspeed',
	'median_doorposition',
]

min_columns = [
	'min_doorforce',
	'min_doorspeed',
	'min_doorposition',
]

max_columns = [
	'max_doorforce',
	'max_doorspeed',
	'max_doorposition',
]
```

```python
# Calculate the mean of time_state_columns
for column in time_state_columns:
    original_data[f'mean_{column}'] = original_data[column].apply(lambda x: np.mean(x))
    original_data[f'median_{column}'] = original_data[column].apply(lambda x: np.median(x))
    original_data[f'min_{column}'] = original_data[column].apply(lambda x: np.min(x))
    original_data[f'max_{column}'] = original_data[column].apply(lambda x: np.max(x))

original_data
```

```python
#Organizing the data based on failure types
failure_groups = original_data.groupby(failure_columns)
```

```python
failure_groups_meta = pd.DataFrame(
	[group_values for group_values, _ in failure_groups],
	columns = failure_columns,
)
failure_groups_meta.to_parquet('../../data/preprocessed_tabular/group_meta.parquet')
failure_groups_meta
```

Saving the data in separate files based on failure types

```python
def save_groups():
	ref_values = list(failure_groups_meta.itertuples(index=False, name=None))

	for (group_index, (values, group)) in enumerate(failure_groups):
		assert(ref_values[group_index] == values)
		g = group[environment_columns + static_state_columns + mean_columns + median_columns + min_columns + max_columns]
		
		g.to_parquet(f'../../data/preprocessed_tabular/group_{group_index}.parquet')

save_groups()
```
